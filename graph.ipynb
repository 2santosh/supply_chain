```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
import numpy as np
import pandas as pd


```

```python
x=np.linspace(0,10,100)
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot()
```

```python
# green star markers with a dashed line
ax.plot(x,np.sin(x),'*--g',label='sin(x)')
# blue circle markers with a solid line
ax.plot(x,np.sin(x+1), 'o-b',label='sin(x+1)')
#
ax.plot(x,np.sin(x+2), ':r',label='sin(x+2)')
#magenta x markers with a dash-dot line
ax.plot(x,np.sin(x+3), 'x-.m',label='sin(x+3)')
#use ax.sert() to adjust a few properties together
ax.set(xlim=(0,11),ylim=(-1.5,1.5),xlabel='x',ylabel='sin(x)',title='Plot Formatting')
#display legend
ax.legend();
plt.show()
```

```python
rng=np.random.RandomState(42)
x=np.arange(50)
y=x+10*rng.randn(50)
colour=rng.randint(0,50,50)
size=np.abs(rng.randn(50))*100
fig=plt.figure(figsize=(10,4))
ax=fig.add_subplot()
axs=ax.scatter(x,y,c=colour,s=size,alpha=0.8)
ax.set(xlabel='x',ylabel='y',title='ScatterPlot')
fig.colorbar(axs)
plt.show()
```

```python
#generate some data
ID=('a','b','c','d','e','f')
cost=(138,252,342,269,654,447)
profit=(260,504,786,542,654,842)
x=np.arange(6)
width=0.3
#plot bar charts
fig,(ax0,ax1)=plt.subplots(2)
ax0.bar(x,cost,width,color='m',label='cost')
ax0.bar(x,profit,width,bottom=cost,color='c',label='profit')
ax0.set(xticks=x,xticklabels=ID,title='Stacked Bar Chart')
ax0.legend(prop={'size':10})

ax1.barh(x,cost,width,color='m',label='cost')
ax1.barh(x+width,profit,width,color='c',label='profit')
ax1.set(yticks=x+width,yticklabels=ID,title='Horizontal Bar Chart')
ax1.legend(prop={'size':10})
fig.tight_layout();
plt.show();
```

```python
#create samples of std. normal distribution
rng=np.random.RandomState(42)
x=rng.randn(1000,3)
colors=['green','red','blue']

n_bins=15
#create a figure an axes
fig, axs =plt.subplots(2,2)
ax0,ax1,ax2,ax3=axs.flatten()
ax0.hist(x,n_bins,density=True,color=colors,label=colors)
ax0.legend(prop={'size':7})
ax0.set_title('bars with legend')
ax1.hist(x,bins=n_bins,density=True,histtype='barstacked')
ax1.set_title('stack step (unfilled)')
ax2.hist(x,n_bins,histtype='step',stacked=True)
ax2.set_title('stack step (unfilled)')
ax3.hist(x,n_bins,histtype='stepfilled',alpha=0.7,stacked=True)
ax3.set_title('stack step (filled)')
fig.tight_layout()
for ax in axs.flat:
    ax.label_outer()
plt.show()
```

```python
#create example data 
product=['webcam','keyboard','mouse','mic','pc']
pct=[20,25,30,10,15]
#only 'explode' the 3rd slice (i.e. 'mouse')
explode=(0,0,0.1,0,0)
#get some colour sheme
cmap=plt.get_cmap('Set2')
colors=cmap(np.arange(5))
#plot pie chart
fig,axs =plt.subplots(1,2,figsize=(8,4))
axs[0].pie(pct,explode=explode,colors=colors,labels=product,
           autopct='%1.1f%%',shadow=True)
axs[0].set_title('Pie chart with explode',fontsize=14)
axs[1].pie(pct,colors=colors,autopct='%1.1f%%',shadow=True)
axs[1].legend(product,title='Products',loc='center left',bbox_to_anchor=(1,0,0.5,1));
plt.show()
```
