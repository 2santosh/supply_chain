```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline

import numpy as np
import pandas as pd
```

### Add Axes on a Figure

```python
# First create some toy data:
x = np.linspace(0, 10, 100)

# Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))

ax.set_title('A single plot', {'fontsize':14});
plt.show()
```

```python
# Create vertically stacked subplots, i.e. ncols=1
fig, axs = plt.subplots(2)
axs[0].plot(x, np.sin(x))
axs[1].plot(x, -np.sin(x))

# Set titles
fig.suptitle('Main Figure Title', fontsize=12)
axs[0].set_title('Axes 0 Title', fontsize=10)
axs[1].set_title('Aces 1 Title', fontsize=10);
```

```python
# We can make the plots neater by sharing X axis
fig, axs = plt.subplots(2, sharex=True)

axs[0].plot(x, np.sin(x))
axs[1].plot(x, -np.sin(x))

# Set titles
fig.suptitle('Main Figure Title', fontsize=12)
axs[0].set_title('Axes 0 Title', fontsize=10)
axs[1].set_title('Aces 1 Title', fontsize=10);
```

```python
# Create a 2D grid subplots
fig, axs = plt.subplots(2, 2)
fig.suptitle('Main Title', fontsize=14)

# Access each subplot using normal indexing for 2D array
axs[0, 0].plot(x, np.sin(x), 'g', linewidth=2)
axs[0, 0].set_title('Axes [0,0] Title')

axs[0, 1].plot(x, np.cos(x), '--r', linewidth=3)
axs[0, 1].set_title('Axes [0,1] Title')

axs[1, 0].plot(x, -np.sin(x), '-.b', linewidth=2)
axs[1, 0].set_title('Axes [1,0] Title')

axs[1, 1].plot(x, -np.cos(x), ':y', linewidth=3)
axs[1, 1].set_title('Axes [1,1] Title')

# Access each subplot using loop
for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots 
# and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
# Alternatively, use tuple-unpacking to assign all subplots to dedicated variables:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Main Title', fontsize=14)

ax1.plot(x, np.sin(x), 'g', linewidth=2)
ax1.set_title('Axes1 Title')

ax2.plot(x, np.cos(x), '--r', linewidth=3)
ax2.set_title('Axes2 Title')

ax3.plot(x, -np.sin(x), '-.b', linewidth=2)
ax3.set_title('Axes3 Title')

ax4.plot(x, -np.cos(x), ':y', linewidth=3)
ax4.set_title('Axes4 Title')

# To get a list of axes in the figure
for ax in fig.get_axes():
    ax.set(xlabel='x-label', ylabel='y-label')
    ax.label_outer()
```

```python
# Choose a different seaborn style
sns.set_style('ticks')

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True,
                        gridspec_kw={'hspace': 0, 'wspace': 0})

# unpack the axes
(ax1, ax2), (ax3, ax4) = axs

# add a main title
fig.suptitle('Main Title', fontsize=14)

ax1.plot(x, np.sin(x), 'g', linewidth=2)
ax2.plot(x, np.cos(x), '--r', linewidth=3)
ax3.plot(x, -np.sin(x), '-.b', linewidth=2)
ax4.plot(x, -np.cos(x), ':y', linewidth=3);
```

```python
# reset sns to default
sns.set()
# create a figure
fig = plt.figure()
fig.suptitle('Main Title', fontsize=14)

# add subplots
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(2,2,2)  #equivalent but more general
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(x, np.sin(x), 'g', linewidth=2)
ax2.plot(x, np.cos(x), '--r', linewidth=3)
ax3.plot(x, -np.sin(x), '-.b', linewidth=2)
ax4.plot(x, -np.cos(x), ':y', linewidth=3);

for ax in fig.get_axes():
    ax.set(xlabel='x-label', ylabel='y-label')
    ax.label_outer()
```

Alternatively, use plt.subplot() method, which is a wrapper for Figure.add_subplot()

```python
ax1=plt.subplot(2, 2, 1)
ax1.plot(x, np.sin(x), 'g', linewidth=2)

# add a subplot with no frame
ax2=plt.subplot(222, sharey=ax1, frameon=False)
ax2.plot(x, np.cos(x), '--r', linewidth=3)

# add a polar subplot
ax3=plt.subplot(223, projection='polar')
ax3.plot(x, -np.sin(x), '-.b', linewidth=2)

# add a red subplot that shares the x-axis with ax1
ax4=plt.subplot(224, sharex=ax1, facecolor='c')
ax4.plot(x, -np.cos(x), ':y', linewidth=3)

#delete ax3 from the figure
plt.delaxes(ax3)

# add ax3 to the figure again
plt.subplot(ax3);
```

### Line plot

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot()

# green star markers with a dashed line
ax.plot(x, np.sin(x), '*--g', label='sin(x)')

# blue circle markers with a solid line
ax.plot(x, np.sin(x+1), 'o-b', label='sin(x+1)')

# red dotted line
ax.plot(x, np.sin(x+2), ':r', label='sin(x+2)')

# magenta x markers with a dash-dot line 
ax.plot(x, np.sin(x+3), 'x-.m', label='sin(x+3)') 

# use ax.set() to adjust a few properties together
ax.set(xlim=(0, 11), ylim=(-1.5, 1.5),
       xlabel='x', ylabel='sin(x)', 
       title='Plot Formatting')

# display legend
ax.legend();
```

### Scatter plot

```python
# Create some random data
rng = np.random.RandomState(42)

x = np.arange(50)
y = x + 10 * rng.randn(50)

colour = rng.randint(0, 50, 50)
size = np.abs(rng.randn(50)) * 100   
            
# scatter plot
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot()
axs = ax.scatter(x, y, c=colour, s=size, alpha=0.8)
ax.set(xlabel='x', ylabel='y', title='Scatter Plot')

# show colour scale
fig.colorbar(axs); 
```

```python
# or simply use plt.scatter()
plt.figure(figsize=(10,4))
plt.scatter(x, y, c=colour, s=size, alpha=0.5, cmap='viridis')
plt.colorbar();
plt.show()
```

Alternatively, we can use plot(), which is faster, but scatter() give us more flexibility in formatting markers. 

```python
# using plot() for same size or colour scatters
plt.plot(x, y, 'og'); # green circlr markers
plt.show()
```

### Bar chart

```python
# generate some data
ID = ('a','b','c','d','e','f')
cost = (138,252,342,269,654,447)
profit = (260, 504, 786, 542, 654, 842)
x = np.arange(6)
width = 0.3

# plot bar charts
fig, (ax0,ax1) = plt.subplots(2)

ax0.bar(x, cost, width, color='m',label='cost')
ax0.bar(x, profit, width, bottom=cost, color='c',label='profit')
ax0.set(xticks=x, xticklabels=ID, title='Stacked Bar Chart')
ax0.legend(prop={'size': 10})

ax1.barh(x, cost, width, color='m', label='cost')
ax1.barh(x+width, profit, width, color='c', label='profit')
ax1.set(yticks=x+width, yticklabels=ID, title='Horizontal Bar Chart')
ax1.legend(prop={'size': 10})

fig.tight_layout();
```

### Histogram

```python
# example data, random standard normal distribution
x = np.random.randn(10000)  
n_bins = 50

plt.hist(x, n_bins, density=1, facecolor="green", alpha=0.5)
plt.xlabel(r"Description of $x$ coordinate (units)")
plt.ylabel(r"Description of $y$ coordinate (units)")
plt.title("Histogram ");
```

```python
# seaborn can plot both a histogram and 
#a continuous density estimate simultaneously
sns.distplot(x,bins=n_bins,color='g');
```

```python
# create samples of std. normal distribution
rng = np.random.RandomState(42)
x = rng.randn(1000, 3)

colors = ['green', 'red', 'blue']
n_bins = 15

# create a figure and axes
fig, axs = plt.subplots(2,2)
ax0, ax1, ax2, ax3 = axs.flatten()

ax0.hist(x, n_bins, density=True, color=colors, label=colors)
ax0.legend(prop={'size': 7})
ax0.set_title('bars with legend')

ax1.hist(x, bins=n_bins, density=True, histtype='barstacked')
ax1.set_title('stacked bar')

ax2.hist(x, n_bins, histtype='step', stacked=True)
ax2.set_title('stack step (unfilled)')

ax3.hist(x, n_bins, histtype='stepfilled', alpha=0.7, stacked=True)
ax3.set_title('stack step (filled)')

fig.tight_layout()
for ax in axs.flat:
    ax.label_outer()
plt.show()
```

### Pie chart

```python
# create example data
product = ['webcam', 'keyboard', 'mouse', 'mic', 'PC']
pct = [20, 25, 30, 10, 15]

# only 'explode' the 3rd slice (i.e.'mouse')
explode = (0, 0, 0.1, 0, 0) 

# get some colour scheme
cmap = plt.get_cmap('Set2')
colors = cmap(np.arange(5))

# plot pie chart
fig, axs = plt.subplots(1,2, figsize=(8,4))
axs[0].pie(pct, explode=explode, colors=colors, labels=product, 
        autopct='%1.1f%%', shadow=True)
axs[0].set_title('Pie chart with explode',fontsize=14)

axs[1].pie(pct, colors=colors, autopct='%1.1f%%', shadow=True)
axs[1].set_title('Pie chart with legend',fontsize=14)
axs[1].legend(product, title='Products',loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1));
```

### Boxplot

```python
rng = np.random.RandomState(42)

# create some random data
spread = rng.rand(50) * 100
center = np.ones(30) * 50
flier_high = rng.rand(5) * 100 + 50
flier_low = rng.rand(5) *100 - 50
data = np.concatenate((spread, center, flier_high, flier_low))

# plot boxplots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, figsize=(8,4))

ax1.boxplot(data, flierprops={'markerfacecolor': 'g'})
ax1.set_title('Basic boxplot with outliers')

ax2.boxplot(data, vert=False, showfliers=False)
ax2.set_title('Horizontal boxplot w/o outliers')

ax3.boxplot(data, notch=True,        # notch shape
                  vert=True)         # vertical box alignment
ax3.set_title('Notched boxplot')

ax4.boxplot(data, notch=True,        # notch shape
                  vert=False,        # horizontal box alignment
                  patch_artist=True) # fill with color
ax4.set_title('Notched colour box')

fig.tight_layout();
```

### Seaborn examples

```python
# load tips data
tips = sns.load_dataset("tips")
tips.head()
```

**Relational plots**

```python
# scatter plot
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips);
plt.show();
```

```python
# set hue to 'size' column
sns.relplot(x='total_bill', y='tip', hue='size', data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', size='size', data=tips);
```

```python
# line plot
sns.relplot(x='total_bill', y='tip', hue='smoker',
            kind='line',height=4, aspect=2, data=tips);
```

**Categorical plots**

```python
sns.catplot(x='day', y='total_bill', hue='sex', data=tips);
```

```python
sns.catplot(x="smoker", y="tip", hue='sex', data=tips);
```

Note: You can use sns.boxplot()for creating a boxplot, however let's try sns.catplot() here

```python
# plot boxplot using catplot()
cat = sns.catplot(x='total_bill', y='day', hue='time',
                  height=3, aspect=1,kind='box', 
                  legend=False, palette='Set2',data=tips)

cat.fig.set_size_inches(10, 4)
cat.add_legend(title='Meal Time')
cat.set_axis_labels('Total bill ($)')
cat.set(xlim=(0, 50), yticklabels=['Thur', 'Fri', 'Sat', 'Sun'])

plt.setp(cat.ax.get_yticklabels(), rotation=30);
```

**Linear Regression Models**

```python
fig, ax=plt.subplots(figsize=(8,4))
sns.regplot(x='total_bill', y='tip', data=tips, ax=ax);
```

While regplot() always shows a single relationship, lmplot() combines regplot() with FacetGrid to provide an easy interface to show a linear regression on faceted plots that allow you to explore interactions with up to three additional categorical variables.

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker',
           col='time', row='sex', data=tips);
```

**Heatmap**

```python
# set a new style
sns.set(style="white")

# generate some random data
rng = np.random.RandomState(46)
df = pd.DataFrame(data=rng.normal(size=(100, 20)),
                  columns=np.arange(20))

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Create a figure and an Axes
fig, ax = plt.subplots(figsize=(12, 8))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
